```python
matching_current_compamy = {}
for code, name in zip(
    local_current_company['업종대분류코드'], 
    local_current_company['업종대분류명'],
):
    try:
        float(code)
    except:
        matching_current_compamy[code] = {
            'name': name,
            'mid_code': {}
        }
matching_current_compamy
```

```python
for code, mid_code, mid_name in zip(
    local_current_company['업종대분류코드'], 
    local_current_company['업종중분류코드'],
    local_current_company['업종중분류명']
):
    try:
        float(code)
    except:
        if code == mid_code[:1]:
            matching_current_compamy[code]['mid_code'][mid_code] = mid_name
matching_current_compamy
```
